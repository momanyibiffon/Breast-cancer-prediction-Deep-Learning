```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('datasets/breast_cancer_wisconsin.csv', index_col=False)
```

```python
data.head(5)
```

```python
data['diagnosis'].unique()
```

```python
data.describe().T
```

```python
# check for null data
data.isnull().sum()
```

```python
# renaming diagnosis to label
data = data.rename(columns={'diagnosis':'Label'})
```

```python
# checking data tyoes
data.dtypes
```

```python
# distribution of data
data['Label'].value_counts()
```

```python
# ploting the dataset
sns.countplot(x='Label', data=data)
```

```python
# defining the dependent variables
y = data['Label'].values # converting to numpy array
np.unique(y)
```

```python
# encoding categorical data from text (B and M) to integers (0 and 1)
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
Y = labelencoder.fit_transform(y) # M = 1, B = 0
```

```python
# labels after encoding are 0 and 1
np.unique(y), np.unique(Y)
```

```python
# define x and normalize/scale the values

# define independent variables and dron label  and id columns

X = data.drop(labels=['Label'], axis=1)
```

```python
X.describe().T
```

```python
# looking at the mean, its not well distributed, therefore, we need to scale our data between 0 and 1
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
# features after scaling
X
```

```python
X.shape
```

```python
# split data into training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### Deep Learning

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
```

```python
# our model
model = Sequential()
model.add(Dense(16, input_dim=30, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1))
model.add(Activation('sigmoid'))

# compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
# fitting the model
history = model.fit(X_train, y_train, verbose=1, epochs=100, batch_size=64, validation_data=(X_test, y_test))
```

```python
# plotting training and validatyion accuracy abd loss at each epoch

# loss at each epoch
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss)+1)
plt.plot(epochs, loss, 'g', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# accuracy at each epoch
accuracy = history.history['accuracy']
val_acc = history.history['val_accuracy']
plt.plot(epochs, accuracy, 'g', label='Training accuracy')
plt.plot(epochs, val_acc, 'r', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# prediction of the test set results
y_pred = model.predict(X_test)
# setting a threshold so that if y_pred < 0.5 = 0 and if y_pred > 0.5 = 1
y_pred = (y_pred > 0.5)
```

```python
# creating a confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```
